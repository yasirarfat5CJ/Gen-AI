```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import SnowballStemmer

# Example paragraph
paragraph = "Dr. A.P.J. Abdul Kalam was a great scientist and former President of India. He inspired millions with his speeches."

# Sentence tokenization
sentences = sent_tokenize(paragraph)

# Initialize Snowball Stemmer
stemming = SnowballStemmer("english")

# Apply stopword removal + stemming
for i in range(len(sentences)):
    words = word_tokenize(sentences[i])  # tokenize into words
    words = [stemming.stem(word) for word in words if word.lower() not in stopwords.words('english')]
    sentences[i] = ' '.join(words)

print(sentences)

```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import SnowballStemmer, WordNetLemmatizer
from nltk.corpus import stopwords

# Download necessary resources (run only once)
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('omw-1.4')

paragraph = """My dear young friends, dream! Dream big, because dreams transform into thoughts,
 and thoughts result in action. You must have a goal in life â€” something that drives you every single day.
   Never give up, no matter how many times you fail. Failure is not the opposite of success; it is a part of success. 
   Each failure teaches you something new. Remember, great things are not achieved by people who give up, 
   but by those who keep trying with faith and hard work.
 If you have courage, determination, and a pure heart, then nothing is impossible for you or for our nation.
 """

# Initialize tools
stemmer = SnowballStemmer('english')
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Sentence tokenization
sentences = sent_tokenize(paragraph)

for i, sentence in enumerate(sentences):
    # Word tokenization
    words = word_tokenize(sentence)
    # Remove stopwords
    words = [word for word in words if word.lower() not in stop_words and word.isalpha()]
    # POS tagging
    pos_tags = nltk.pos_tag(words)
    
    # Apply stemming and lemmatization
    stems = [stemmer.stem(word) for word in words]
    lemmas = [lemmatizer.lemmatize(word) for word in words]
    
    print(f"\nSentence {i+1}: {sentence}")
    print("POS Tags:", pos_tags)
    print("Stemmed Words:", stems)
    print("Lemmatized Words:", lemmas)

```
