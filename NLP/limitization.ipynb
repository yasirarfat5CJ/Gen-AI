```python
#WordnetLemmatizer

#its a technique like stemming the o/p will get lemmatization iscalled lemma  which is a root word rather then a root stem

from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()

lemmatizer.lemmatize("going",pos='v')
words = ["running", "studies", "easily", "flies", "working","sportingly","goes","history"]  # example word list
for word in words:
    print(lemmatizer.lemmatize(word,pos='v'))


##use cases 
## Q & a
## Text summarization
#3 chatbot
```
