```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import SnowballStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
# Initialize tools
stemmer = SnowballStemmer('english')
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Sentence tokenization
sentences = sent_tokenize(para)

for i, sentence in enumerate(sentences):
    # Word tokenization
    words = word_tokenize(sentence)
    # Remove stopwords
    words = [word for word in words if word.lower() not in stop_words and word.isalpha()]
    # POS tagging
    pos_tags = nltk.pos_tag(words)
    
    # Apply stemming and lemmatization
    stems = [stemmer.stem(word) for word in words]
    lemmas = [lemmatizer.lemmatize(word) for word in words]
    
    print(f"\nSentence {i+1}: {sentence}")
    print("POS Tags:", pos_tags)
    
```

```python
##NAME ENTITY RECOGNIZATION
import nltk

words = "Elon Musk founded SpaceX in 2002, based in California."
# Download necessary NLTK data files
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')
# nltk.download('maxent_ne_chunker')
# nltk.download('words')



# Tokenize
words = nltk.word_tokenize(words)
print("Tokens:", words)

# Part-of-Speech Tagging
tagged = nltk.pos_tag(words)
print("\nPOS Tags:", tagged)

# Named Entity Recognition (NER)
tree = nltk.ne_chunk(tagged)
print("\nNamed Entity Recognition Tree:")
print(tree)

# To visualize it (optional)
# tree.draw()

```
