```python
##pasword strength checker
def checkpassword(password):
    if len(password)<8:
        return "password is weak"
    if not any(char.isdigit() for char in password):
        return "password is weak."
    if not any(char.islower() for char in password):
        return "password  must conatin any  lower case"
    if not any(char.isupper() for char in password):
        return "password must contain any upper case"
    if not any(char in '!@#$%^&*()_+' for char in password):
        return "password must conatin special character"
    return "passwoord is strong"
checkpassword("Arfat123!")
```

```python
##Modeling a bank account

class Bank:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} is deposited. New balance: {self.balance}")
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance = self.balance - amount
            print(f'{amount} is withdrawn. New Balance is {self.balance}')
    def get_balance(self):
        return self.balance

account = Bank("me", 5000)
print(account.get_balance())
account.deposit(2000)
print(account.get_balance())
account.withdraw(8000)

    
```

```python

class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print("animal is speaking")

class Pet:
    def __init__(self, name, owner):
        super().__init__(name)   # call Animal's __init__
        self.owner = owner
    
    def speak(self):
        print("pet is speaking")

class Dog(Pet, Animal):   # Order matters for MRO
    def __init__(self, name, owner):
        super().__init__(name, owner)   # uses MRO to call Pet -> Animal
       
    def speak(self):
        print("dog is speaking")

dog1 = Dog("Buddy", "yasir")
dog1.speak()
print(dog1.owner)
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y   # fix tuple issue

    def __str__(self):   # for user-friendly print
        return f"Vector({self.x}, {self.y})"

    def __repr__(self):  # for debugging / interactive mode
        return f"Vector({self.x}, {self.y})"


v1 = Vector(2, 3)
v2 = Vector(4, 5)

print(v1 + v2)        # Vector(6, 8)
print(v2 - v1)        # Vector(2, 2)
print(v1 == v2)       # False
print(v1 == Vector(2, 3))  # True
```
