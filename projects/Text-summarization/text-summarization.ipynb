```python
import os 
from dotenv import load_dotenv
load_dotenv()

api_key=os.getenv("GROQ_API_KEY")

```

```python
from langchain_groq import ChatGroq

llm=ChatGroq(groq_api_key=api_key,model_name="llama-3.3-70b-versatile")
```

```python
from langchain_core.messages import AIMessage,HumanMessage,SystemMessage
```

```python
speech="""
My dear young friends,

I am delighted to be with you today and see the bright, confident faces of the future of India. When I look at you, I do not see students alone — I see scientists, doctors, teachers, entrepreneurs, leaders, and innovators who will shape the destiny of our nation.

Dream, dream, dream.
Dreams transform into thoughts, and thoughts result in action.

Each one of you has a unique capability and immense potential. Never compare yourself with others; instead, compare yourself with who you were yesterday. Excellence is a continuous process, not an accident.

India is a nation with a billion people, yet it is also a nation with a billion dreams. If we want India to become a developed nation, we must ignite young minds. Young minds are the most powerful resource on the earth, above the earth, and under the earth.

You will face problems in life. Do not fear them. Problems are common, but attitude makes the difference. """
```

```python
chat_message=[
    SystemMessage(content="You are an expert in summarizing content"),
    HumanMessage(content=f"please provide a short and concise summary of the given text\n:{speech}")
]
```

```python
llm.get_num_tokens(speech)
```

```python
llm.invoke(chat_message)
```

```python
##technique2  for summarization

from langchain_classic.chains import LLMChain
from langchain_core.prompts import PromptTemplate

generic_prompt="""
Write  a summary of the following speech:
speech:{speech}
Translate the precise summary to {language}
"""
prompt=PromptTemplate(
    input_variables=['speech','language'],
    template=generic_prompt
)
prompt
```

```python
complete_prompt=prompt.format(speech=speech,language="hindi")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
summary=llm_chain.invoke({"speech":speech,"language":"hindi"})
```

```python
##========1)Stuff document chain text summarization technique

## all documents will combine then it send to prompt template it goes to  LLM then LLm  produues o/p
##promblem  : if docs becomes very  big then we cannot send directly


##=======2)Map Reduce summarization technique

## in this first documents wull be divided in to chuks then passed to prompt template then LLM then we get summary1 summary2 summary3....  finally we combine and get the final summary
```

```python
## Implementation
from  langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("Attention.pdf")
docs=loader.load()
docs
```

```python
template="""
Give an concise  sand summary of the given content
input:{text}
"""
prompt=PromptTemplate(input_variables=['text'],template=template)
```

```python
from langchain_classic.chains import load_summarize_chain

chain=load_summarize_chain(llm,chain_type='stuff',prompt=prompt,verbose=True)
```

```python
chain.invoke(docs)
```

```python
##map reduce is used to summarize the large documents
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
final_docs=splitter.split_documents(docs)
final_docs
```

```python
chunks_prompt="""
Please summarize the below speech 
Speech:{text}
summary:
"""
map_prompt_template=PromptTemplate(input_variables=['text'],template=chunks_prompt)

```

```python
summary_chain=load_summarize_chain(llm,chain_type='map_reduce',map_prompt=map_prompt_template,combine_prompt=map_prompt_template,verbose=True)
summary_chain.invoke(final_docs)
```

```python
###======3)Refine  chain summarization

#it as an extension op mapreduer summarization we will give the res of first to the input of secound docs and combined  and give to the prompt and produce the output
```

```python
refine_chain=load_summarize_chain(llm,chain_type="refine",verbose=True)
refine_chain.invoke(final_docs)
```
