```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('COA UNIT-I.pdf')
docs=loader.load()
docs
```

```python
# How to recursive split text 
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
# docs is already a list of Document objects; use split_documents to handle Documents directly
final_docs = text_splitter.split_documents(docs)
final_docs
```

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader('speech.txt')

docs=loader.load()
docs
```

```python
speech=""
with open('speech.txt') as f:
    speech=f.read()
speech

text_splitter=RecursiveCharacterTextSplitter(chunk_size=20,chunk_overlap=10)
text=text_splitter.create_documents([speech])
text
```

```python
## Character splitter
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=20,chunk_overlap=10)
text_splitter.split_documents(docs)
```

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

# HTML content (must be assigned to a variable)
html_string = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>LangChain HTMLTextSplitter Demo</title>
</head>
<body>

    <h1>LangChain HTML Text Splitter</h1>

    <p>
        LangChain is a framework designed to build applications powered by large language models.
        It helps developers connect LLMs with external data sources.
    </p>

    <br>

    <h2>Why Use HTMLTextSplitter?</h2>

    <div>
        <p>
            HTMLTextSplitter preserves the structure of HTML documents while splitting text.
            This helps maintain context during document chunking.
        </p>

        <p>
            Unlike normal text splitters, it understands headings and sections.
        </p>
    </div>

</body>
</html>
"""

# Headers to split on
headers_to_split_on = [
    ("h1", "header1"),
    ("h2", "header2"),
]

# Create splitter and split HTML
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
docs = html_splitter.split_text(html_string)
html_string
# Display results
for i, doc in enumerate(docs, 1):
    print(f"\n--- Document {i} ---")
    print("Metadata:", doc.metadata)
    print("Content:", doc.page_content)
```

```python
## SPLIT JSON DATA
import json
import requests

json_data=requests.get("https://api.smith.langchain.com/openapi.json").json()
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
json_spliter=RecursiveJsonSplitter(max_chunk_size=300)

json_chunks=json_spliter.split_json(json_data)
json_chunks
```

```python
docs=json_spliter.create_documents(texts=[json_data])
docs
```
